```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('medical_appointments.csv')
```

```python
data
```

```python
bar_chart = data.drop(columns=['PatientId'])
```

```python
bar_chart
```

```python


# Count values
bar_data = data['NoShow'].value_counts()

# Plot
plt.figure(figsize=(6,4))
plt.bar(bar_data.index, bar_data.values)
plt.title('No Show vs Show Appointments')
plt.xlabel('Appointment Status')
plt.ylabel('Count')
plt.show()
```

```python
data.isnull().sum()
```

```python
# Age → median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Gender → mode
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)

# AppointmentDay → mode
data['AppointmentDay'].fillna(data['AppointmentDay'].mode()[0], inplace=True)

# Diabetes → mode
data['Diabetes'].fillna(data['Diabetes'].mode()[0], inplace=True)

# SMS_Received → mode
data['SMS_Received'].fillna(data['SMS_Received'].mode()[0], inplace=True)
```
