```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('adult_income.csv')
```

```python
categorical_features = data.select_dtypes(include='object').columns
numerical_features = data.select_dtypes(exclude='object').columns

print("Categorical:", list(categorical_features))
print("Numerical:", list(numerical_features))
```

```python


le = LabelEncoder()

data['Education'] = le.fit_transform(data['Education'])
data['Income'] = le.fit_transform(data['Income'])
```

```python
data_encoded = pd.get_dummies(
    data,
    columns=['Workclass', 'Marital_Status', 'Occupation'],
    drop_first=True
)
```

```python


scaler = StandardScaler()

num_cols = ['Age', 'Education_Num', 'Hours_per_week']

data_scaled = data_encoded.copy()
data_scaled[num_cols] = scaler.fit_transform(data_encoded[num_cols])
```

```python
print("Before scaling:\n")
print(data_encoded[num_cols].head())

print("\nAfter scaling:\n")
print(data_scaled[num_cols].head())
```
